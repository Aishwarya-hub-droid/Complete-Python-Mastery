```python
# capitalize()	It capitalizes the first character of the String. This function is deprecated in python3
website = 'www.codeswithpankaj.com'
print(website.capitalize())

# 1. upper() converts the string to uppercase
print(website.upper())

# 2. lower() converts the string to lowercase
print(website.lower())

# 3. partition() returns a tuple
print(website.partition('.'))

# 4. replace() replaces substring inside
print(website.replace('www.','@'))

# 5. find() returns the index of first occurrence of substring
print(website.find('j'))

# 6. rstrip() removes trailing characters
data = 'p4n.in   '
print(data.rstrip(),'this is website -')

# 7. split() splits string from left
print(website.split('.'))

# 8. startswith() checks if string starts with the specified string
print(website.startswith('w'))

# 9. isnumeric() checks numeric characters
data = '90'
print(data.isnumeric())
print(website.isnumeric())

# 10. index() returns index of substring
print(website.index('p'))
```

```python
# Input string
input_string = "p4n.in"

# Length of the string
length = len(input_string)
print(f"Length of the string: {length}")

# Convert to lowercase
lowercase_string = input_string.lower()
print(f"Lowercase: {lowercase_string}")

# Convert to uppercase
uppercase_string = input_string.upper()
print(f"Uppercase: {uppercase_string}")

# Count the number of occurrences of a substring
substring = "l"
count = input_string.count(substring)
print(f"Number of '{substring}' in the string: {count}")

# Find the index of a substring (first occurrence)
substring = "World"
index = input_string.find(substring)
if index != -1:
    print(f"Index of '{substring}': {index}")
else:
    print(f"'{substring}' not found in the string.")

# Replace a substring
old_substring = "World"
new_substring = "Python"
new_string = input_string.replace(old_substring, new_substring)
print(f"String after replacement: {new_string}")

# Split the string into a list of words
words = input_string.split()
print(f"List of words: {words}")
```

```python
# Sample input string
input_string = "Welcome to Codes With Pankaj @codeswithpankaj.com"

# capitalize()
capitalized_string = input_string.capitalize()
print("capitalize():", capitalized_string)

# casefold()
casefolded_string = input_string.casefold()
print("casefold():", casefolded_string)

# center()
centered_string = input_string.center(50, '*')
print("center():", centered_string)

# count()
count_occurrences = input_string.count("is")
print("count():", count_occurrences)

# encode()
encoded_string = input_string.encode()
print("encode():", encoded_string)

# endswith()
ends_with = input_string.endswith("Testing")
print("endswith():", ends_with)

# expandtabs()
expanded_tabs = "Tab\tSeparated\tString".expandtabs(4)
print("expandtabs():", expanded_tabs)

# find()
find_position = input_string.find("is")
print("find():", find_position)

# format()
formatted_string = "Hello, {}!".format("John")
print("format():", formatted_string)

# format_map()
formatted_map = "Hello, {name}!".format_map({"name": "Alice"})
print("format_map():", formatted_map)

# index()
index_position = input_string.index("codes")
print("index():", index_position)

# isalnum()
is_alphanumeric = input_string.isalnum()
print("isalnum():", is_alphanumeric)

# isalpha()
is_alpha = input_string.isalpha()
print("isalpha():", is_alpha)

# isascii()
is_ascii = input_string.isascii()
print("isascii():", is_ascii)

# isdecimal()
is_decimal = "12345".isdecimal()
print("isdecimal():", is_decimal)

# isdigit()
is_digit = "12345".isdigit()
print("isdigit():", is_digit)

# isidentifier()
is_identifier = "my_variable".isidentifier()
print("isidentifier():", is_identifier)

# islower()
is_lower = input_string.islower()
print("islower():", is_lower)

# isnumeric()
is_numeric = "12345".isnumeric()
print("isnumeric():", is_numeric)

# isprintable()
is_printable = input_string.isprintable()
print("isprintable():", is_printable)

# isspace()
is_space = "   ".isspace()
print("isspace():", is_space)

# istitle()
is_title = input_string.istitle()
print("istitle():", is_title)

# isupper()
is_upper = input_string.isupper()
print("isupper():", is_upper)

# join()
my_list = ["Hello", "World", "Python"]
joined_string = "-".join(my_list)
print("join():", joined_string)

# ljust()
left_justified_string = input_string.ljust(50, '-')
print("ljust():", left_justified_string)

# lower()
lowercase_string = input_string.lower()
print("lower():", lowercase_string)

# lstrip()
left_stripped_string = "   Sample String   ".lstrip()
print("lstrip():", left_stripped_string)

# maketrans() and translate()
translation_table = input_string.maketrans("aeiou", "12345")
translated_string = input_string.translate(translation_table)
print("translate():", translated_string)

# partition()
partitioned_string = input_string.partition("is")
print("partition():", partitioned_string)

# replace()
replaced_string = input_string.replace("com", "@p4n.in")
print("replace():", replaced_string)

# rfind()
rfind_position = input_string.rfind("com")
print("rfind():", rfind_position)

# rindex()
rindex_position = input_string.rindex("com")
print("rindex():", rindex_position)

# rjust()
right_justified_string = input_string.rjust(50, '-')
print("rjust():", right_justified_string)

# rpartition()
rpartitioned_string = input_string.rpartition("com")
print("rpartition():", rpartitioned_string)

# rsplit()
rsplit_list = input_string.rsplit(" ", 2)
print("rsplit():", rsplit_list)

# rstrip()
right_stripped_string = "   Sample String   ".rstrip()
print("rstrip():", right_stripped_string)

# split()
split_list = input_string.split(" ")
print("split():", split_list)

# splitlines()
split_lines = "codes\nwith\npankaj".splitlines()
print("splitlines():", split_lines)

# startswith()
starts_with = input_string.startswith("codes")
print("startswith():", starts_with)

# strip()
stripped_string = "   Sample String   ".strip()
print("strip():", stripped_string)

# swapcase()
swapped_case_string = input_string.swapcase()
print("swapcase():", swapped_case_string)

# title()
title_case_string = input_string.title()
print("title():", title_case_string)

# upper()
uppercase_string = input_string.upper()
print("upper():", uppercase_string)

# zfill()
zfilled_string = "42".zfill(5)
print("zfill():", zfilled_string)
```
