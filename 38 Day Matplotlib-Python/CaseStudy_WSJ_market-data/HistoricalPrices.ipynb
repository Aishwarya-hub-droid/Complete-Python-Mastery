```python
import pandas as pd
data = pd.read_csv('HistoricalPrices.csv')
data.head()
```

```python
data = data.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values(by = 'Date')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
plt.plot(data['Date'], data['Open'])
plt.show()
```

```python
plt.plot(data['Date'], data['Open'])
plt.plot(data['Date'], data['Close'])
plt.plot(data['Date'], data['High'])
plt.plot(data['Date'], data['Low'])
plt.show()
```

```python
plt.plot(data['Date'], data['Open'],label ='Open')
plt.plot(data['Date'], data['Close'],label = 'Close')
plt.plot(data['Date'], data['High'],label ='High')
plt.plot(data['Date'], data['Low'],label = 'Low')
plt.legend()
plt.show()
```

```python
# Import the calendar package 
from calendar import month_name
# Order by months by chronological order
data['Month'] = pd.Categorical(data['Date'].dt.month_name(), month_name[:])

# Group metrics by monthly averages
monthly = data \
    .groupby('Month') \
    .mean() \
    .reset_index()

monthly.head(6)
```

```python
plt.bar(monthly['Month'], height = monthly['Close'])
plt.show()
```

```python
plt.barh(monthly['Month'], width = monthly['Close'])
plt.show()
```

```python
plt.scatter(data['Open'], data['Close'])
plt.show()
```

```python
z = np.polyfit(data['Open'], data['Close'], 1)
p = np.poly1d(z)


plt.scatter(data['Open'], data['Close'])
plt.plot(data['Open'], p(data['Open']))
plt.show()
```

```python
plt.scatter(data['High'], data['Low'])
plt.show()
```

```python
plt.plot(data['Date'], data['Open'], color = 'black')
plt.plot(data['Date'], data['Close'], color = 'red')
plt.plot(data['Date'], data['High'], color = 'green')
plt.plot(data['Date'], data['Low'], color = 'pink')
plt.show()
```

```python
plt.bar(monthly['Month'], height = monthly['Close'], color = ['blue', 'green', 'yellow', 'red', 'pink', 'black'])
plt.show()
```
