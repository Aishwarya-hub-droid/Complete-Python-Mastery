```python
# import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# simple example

x = np.linspace(-5, 5, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, label='Normal Distribution (mean=0, std=1)')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show() 
```

```python
# simple example 

data = np.array([1, 2, 3, 4, 5])
mean = np.mean(data)

std_dev = np.std(data)
print(f"Mean: {mean}, Standard Deviation: {std_dev}")

# make a normal distribution with mean=0 and std=1
# using data from the previous example
y = stats.norm.pdf(data, mean, std_dev)
plt.plot(data, y, label='Normal Distribution (mean=0, std=1)')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

```python
# simple example
# ganerate random data

data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```
