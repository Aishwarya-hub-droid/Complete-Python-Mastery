```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading the TitanicÂ Dataset
df = sns.load_dataset("titanic")
df.head()
```

```python
# Creating a Pairplot to Explore Relationships

sns.pairplot(df, hue="sex", diag_kind="kde")
plt.show()
```

```python
# Creating a Pairplot to Explore Relationships with different parameters
sns.pairplot(df, hue="pclass", diag_kind="hist")
plt.show()
```

```python
# Visualizing Correlation with aÂ Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
# Distribution of Age with a Histogram
sns.histplot(df['age'].dropna(), kde=True, bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```
